```python
import pandas as pd

# Method 1: Create a DataFrame from a dictionary (most common way)
data = {
    "Name": ["Alice", "Bob", "Charlie", "Diana"],
    "Age": [25, 30, 35, 28],
    "City": ["New York", "London", "Tokyo", "Paris"],
    "Salary": [70000, 80000, 92000, 78000]
}

df = pd.DataFrame(data)

# Display the DataFrame
print(df)
```
